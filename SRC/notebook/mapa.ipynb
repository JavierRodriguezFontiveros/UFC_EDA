```python
import folium
from folium.features import DivIcon
```

```python
#Creando diccionario de cordenadas y numero de peleadores
numero_peladores_dict = {
    'Estados Unidos': {'Peleadores': 140, 'Cordenadas': [37.0902, -95.7129]},
    'Brasil': {'Peleadores': 70, 'Cordenadas': [-14.2350, -51.9253]},
    'Rusia': {'Peleadores': 32, 'Cordenadas': [61.5240, 105.3188]},
    'Inglaterra': {'Peleadores': 22, 'Cordenadas': [52.3555, -1.1743]},
    'México': {'Peleadores': 17, 'Cordenadas': [23.6345, -102.5528]},
    'Canadá': {'Peleadores': 17, 'Cordenadas': [56.1304, -106.3468]},
    'Australia': {'Peleadores': 13, 'Cordenadas': [-25.2744, 133.7751]},
    'España': {'Peleadores': 8, 'Cordenadas': [40.4637, -3.7492]},
    'Nueva Zelanda': {'Peleadores': 7, 'Cordenadas': [-40.9006, 174.8860]},
    'China': {'Peleadores': 7, 'Cordenadas': [35.8617, 104.1954]}
}


mapa_mundi = folium.Map(location=[20, 0], zoom_start=2, tiles='cartodb positron')


for idx, (pais, info) in enumerate(numero_peladores_dict.items(), 1): 
    
    Cordenadas = info['Cordenadas']
    Peleadores = info['Peleadores']

    color = 'green' if idx <= 3 else 'red'
    
    folium.CircleMarker(location=Cordenadas,
                        radius=Peleadores / 2,                         
                        popup=f'{pais}: {Peleadores} peleadores',
                        color=color,
                        fill=True,
                        fill_color=color,
                        fill_opacity=0.7).add_to(mapa_mundi)
    
    folium.Marker(location=Cordenadas,
                  icon=DivIcon(icon_size=(12, 12), 
                               icon_anchor=(6, 6),
                               html=f'<div style="font-size: 12px; color: black;">{idx}</div>',)).add_to(mapa_mundi)

legend_html = '''
<div style="
    position: fixed;
    bottom: 50px; left: 50px; width: 250px; height: 275px;
    border:2px solid grey; z-index:9999; font-size:14px;
    background-color:white; padding: 10px;">
    <h2>Países y Peleadores</h2>
    <ol>
        &nbsp
        <li><b>Estados Unidos</b> - 349 </li>
        <li><b>Brasil</b> - 87 </li>
        <li><b>Rusia</b> - 32 </li>
        <li><b>Inglaterra</b> - 22 </li>
        <li><b>México</b> - 17 </li>
        <li><b>Canadá</b>- 17 </li>
        <li><b>Australia</b> - 13 </li>
        <li><b>España</b> - 8 </li>
        <li><b>Nueva Zelanda</b> - 7 </li>
        <li><b>China</b> - 7 </li>
    </ol>
</div>
'''

mapa_mundi.get_root().html.add_child(folium.Element(legend_html))

mapa_mundi.save("mapa_mundi.html")

mapa_mundi
```
